```python
import pandas as pd
import numpy as np

series_1 = pd.Series([1, 2, 3, np.nan, 5, np.nan, 'hello series', [
                     6, 7, 8]], name='series_1', index=['A', '1', '2', 'B', '3', 'C', '4', 'D'])
print('='*30, '原序列', '='*30)
print(series_1)


# 简单 for 循环
print('='*30, '简单for循环', '='*30)
for i in range(len(series_1)):
    print('第{}个元素为：{}，其index为：{}'.format(
        i, series_1.iloc[i], series_1.index[i]))

# for 循环寻找series中的特定类型所有元素
print('='*30, '查找特定类型', '='*30)
goal_type = int
for i in range(len(series_1)):
    if type(series_1.iloc[i]) == goal_type:
        print('第{}个元素为：{}型，其index为：{}'.format(i, goal_type, series_1.index[i]))
    else:
        pass

# for 循环找出空值：np.nan
print('='*30, '找出空值', '='*30)
for i in range(series_1.size):
    # 使用isnan，必须传入array类型，所以要先判断，否则会报错。使用pd.isnull()也是一样要注意输入数据类型
    if type(series_1.iloc[i]) == int or type(series_1.iloc[i]) == float:
        if np.isnan(series_1.iloc[i]):
            print('第{}个元素为空，其index为：{}'.format(i, series_1.index[i]))

# for 循环将所有数值都乘以100倍，如果是np.nan,在pandas计算sum等操作中会默认为0
print('='*30, '数值操作', '='*30)
for i in range(series_1.size):
    if type(series_1.iloc[i]) == int or type(series_1.iloc[i]) == float:
        series_1.iloc[i] = series_1.iloc[i]*100
    else:
        pass
print(series_1)

# for循环找出小于某个阈值的数
print('='*25, '查找小于阈值的数', '='*25)
threshold = 200
for i in range(series_1.size):
    # 也要先判断数据类型哦
    if type(series_1.iloc[i]) == int or type(series_1.iloc[i]) == float:
        if series_1.iloc[i] < threshold:
            print('第{}个元素小于{}阈值，值为：{}'.format(i, threshold, series_1.iloc[i]))
        else:
            pass
    else:
        print('第{}个元素不符合数据类型要求'.format(i, series_1.iloc[i]))

# for 循环更改数据类型：str、int、folat相互转换
# 将int转为float
print('='*30, '更改数据类型', '='*30)
for i in range(series_1.size):
    if type(series_1.iloc[i]) == int:
        series_1.iloc[i] = float(series_1.iloc[i])
    print(type(series_1.iloc[i]))
```

```python
import pandas as pd
import numpy as np

series_1 = pd.Series([1, 2, 3, np.nan, 5, np.nan, 6, 7], name='series_1', index=[
                     'A', '1', '2', 'B', '3', 'C', '4', 'D'])
print('='*30, '原序列', '='*30)
print(series_1)

# 迭代器 items()
print('='*30, '简单迭代器', '='*30)
for index, item in series_1.items():
    print('index为：{},值为：{}'.format(index, item))

# 使用map函数，将所有值都乘以100倍
def multiply(data):
    times = 100
    return data*times


series_1 = series_1.map(multiply)
print('='*30, 'map应用', '='*30)
print(series_1)

# 使用apply函数，将所有值都乘以100
print('='*30, 'apply应用', '='*30)
series_1 = series_1.apply(multiply)
print(series_1)

# lambda函数，将所有值乘以100
print('='*30, 'apply+lambda应用', '='*30)
series_1 = series_1.apply(lambda x: (x*100))
print(series_1)

# 使用apply函数，转换数据类型，将float改为str
def class_trans(data):
    # np.nan为float，不能强制转换为int，转为int时要注意。转为str不必。
    if np.isnan(data):
        return data
    else:
        if type(data) == float:
            return str(data)
        else:
            return data


series_2 = series_1.apply(class_trans)
print('='*20, 'apply函数更改数据类型', '='*20)
for index, item in series_2.items():
    print('index:{},类型为：{}'.format(index, type(item)))

# 缺失值填充
def filled_nan(data):
    if np.isnan(data):
        return -999
    else:
        return data

print('='*25, 'apply填充缺失值', '='*25)
# 先将str类型转为float
series_1 = series_1.apply(filled_nan)
print(series_1)
```

```python
import pandas as pd
import numpy as np

series_1 = pd.Series([1, 2, 3, np.nan, 5, np.nan, 6, 7], name='series_1', index=[
                     'A', '1', '2', 'B', '3', 'C', '4', 'D'])
# 迭代器 items()
print('='*20, '简单迭代器，不需设置循环次数', '='*20)
for index, item in series_1.items():
    print('index为：{},值为：{}'.format(index, item))

# 配合while循环感受next用法
print('='*20, 'next方法在迭代器中的应用', '='*20)

iterator_series = series_1.__iter__()
while True:
    try:
        next_item = next(iterator_series)
        print(next_item)
    # 要判断迭代器是否迭代至最后一个元素！
    except StopIteration:
        print('迭代取值结束')
        # 在while循环中，break很重要哦，没有的话会一直打印：print('迭代取值结束')
        break
```
